```python
#q1
import numpy as np
from scipy import stats

array = np.random.rand(100)

mean = np.mean(array)
median = np.median(array)
variance = np.var(array)

print("mean:",mean)
print("median",median)
print("variance",variance)
```

```python
#q3
import numpy as np
from scipy import linalg

matrix = np.array([[1, 2], [3, 4]])

determinant = linalg.det(matrix)

inverse = linalg.inv(matrix)

eigenvalues = linalg.eigvals(matrix)

print("Determinant:",determinant);
print("Inverse:",inverse);
print("Eigenvalues:",eigenvalues);
```

```python
#q4
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 10)
y = np.sin(x)

interp_func = interpolate.interp1d(x, y, kind='cubic')
x_new = np.linspace(0, 10, 100)
y_new = interp_func(x_new)

plt.plot(x, y, 'o', label='Original data')
plt.plot(x_new, y_new, '-', label='Interpolated data')
plt.legend()
plt.show()
```

```python
#q5
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 500, endpoint=False)
signal_data = np.sin(2 * np.pi * 7 * t) + np.random.randn(500) * 0.5

sos = signal.butter(10, 15, 'lp', fs=500, output='sos')
filtered = signal.sosfilt(sos, signal_data)

plt.plot(t, signal_data, label='Noisy Signal')
plt.plot(t, filtered, label='Filtered Signal')
plt.legend()
plt.show()
```

```python
#q6
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

np.random.seed(0)
months = np.arange(1, 13)
products = ['Product_A', 'Product_B', 'Product_C', 'Product_D']

sales_data = np.random.randint(100, 1000, size=(12, 4))

total_sales_per_product = np.sum(sales_data, axis=0)

average_sales_per_product = np.mean(sales_data, axis=0)

total_sales_per_month = np.sum(sales_data, axis=1)

best_month = np.argmax(total_sales_per_month) + 1
worst_month = np.argmin(total_sales_per_month) + 1

best_product_index = np.argmax(total_sales_per_product)
best_product = products[best_product_index]

print(f"Total sales per product: {total_sales_per_product}")
print(f"Average sales per product: {average_sales_per_product}")
print(f"Best performing month (highest total sales): Month {best_month}")
print(f"Worst performing month (lowest total sales): Month {worst_month}")
print(f"Best performing product: {best_product}")
```

```python
#q7
import numpy as np
from scipy import stats

marks_data = np.array([
    [85, 78, 92, 88],
    [79, 82, 74, 90],
    [90, 85, 89, 92],
    [66, 75, 80, 78],
    [70, 68, 75, 85]
])

students = ['Arin', 'Aditya', 'Chirag', 'Gurleen', 'Kunal']
subjects = ['Math', 'Physics', 'Chemistry', 'English']

total_marks = np.sum(marks_data, axis=1)
average_marks = np.mean(marks_data, axis=1)

subject_wise_average = np.mean(marks_data, axis=0)

top_performer_index = np.argmax(total_marks)
bottom_performer_index = np.argmin(total_marks)

top_performer = students[top_performer_index]
bottom_performer = students[bottom_performer_index]

passing_percentage = np.mean(marks_data >= 40, axis=1) * 100

print(f"Total marks for each student: {dict(zip(students, total_marks))}")
print(f"Average marks for each student: {dict(zip(students, average_marks))}")
print(f"Subject-wise average marks: {dict(zip(subjects, subject_wise_average))}")
print(f"Top performer: {top_performer} with {total_marks[top_performer_index]} marks")
print(f"Bottom performer: {bottom_performer} with {total_marks[bottom_performer_index]} marks")
print(f"Passing percentage for each student: {dict(zip(students, passing_percentage))}")
```

```python
#q8
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

time = np.array([0, 1, 2, 3, 4, 5])
velocity = np.array([2, 3.1, 7.9, 18.2, 34.3, 56.2])

def quadratic(t, a, b, c):
    return a*t**2 + b*t + c

params, _ = curve_fit(quadratic, time, velocity)

fitted_velocity = quadratic(time, *params)

plt.plot(time, velocity, 'o', label='Data')
plt.plot(time, fitted_velocity, '-', label='Fitted Curve')
plt.xlabel('Time (seconds)')
plt.ylabel('Velocity (m/s)')
plt.legend()
plt.show()
```

<!-- Project 1: Tic-Tac-Toe -->
<div style="background-color: #f7f7f7; padding: 20px; border-left: 5px solid #FF9800; margin: 20px 0;">
  <h1 style="color: #F57C00;">Project 1: Tic-Tac-Toe Game üéÆ</h1>
  <hr style="border: 1px solid #FFE0B2;">
  
  <h2>Objective</h2>
  <p>Develop an interactive Tic-Tac-Toe game in Python to reinforce fundamental programming concepts.</p>
  
  <h3>Implementation Requirements:</h3>
  <ul>
    <li>üéØ <strong>Core Functions:</strong>
      <ul>
        <li><code>print_board(board)</code>: Display game board</li>
        <li><code>check_winner(board, player)</code>: Check for win</li>
        <li><code>check_tie(board)</code>: Check for tie</li>
        <li><code>get_player_input(player, board)</code>: Handle moves</li>
        <li><code>play_game()</code>: Main game loop</li>
      </ul>
    </li>
    <li>üé≤ <strong>Game Board:</strong>
      <ul>
        <li>3x3 grid using list</li>
        <li>Positions numbered 1-9</li>
        <li>Players: X and O</li>
      </ul>
    </li>
    <li>üîç <strong>Win Conditions:</strong>
      <ul>
        <li>Rows: [0,1,2], [3,4,5], [6,7,8]</li>
        <li>Columns: [0,3,6], [1,4,7], [2,5,8]</li>
        <li>Diagonals: [0,4,8], [2,4,6]</li>
      </ul>
    </li>
    <li>‚ö° <strong>Features:</strong>
      <ul>
        <li>Input validation</li>
        <li>Alternate turns</li>
        <li>Win/Tie detection</li>
        <li>Play again option</li>
      </ul>
    </li>
  </ul>
</div>

```python
def print_board(board):
    print("\n")
    for row in board:
        print("| " + " | ".join(row) + " |")
    print("\n")

def check_winner(board, player):
    # Check rows, columns, and diagonals for a winning combination
    for i in range(3):
        if all([spot == player for spot in board[i]]) or all([board[j][i] == player for j in range(3)]):
            return True
    # Check diagonals
    if board[0][0] == board[1][1] == board[2][2] == player or board[0][2] == board[1][1] == board[2][0] == player:
        return True
    return False

def check_tie(board):
    return all([spot != " " for row in board for spot in row])

def get_player_input(player, board):
    while True:
        try:
            # Get user input and split it into row and column
            row, col = map(int, input(f"Player {player}, enter your move (row and column, e.g., 1 2): ").split())
            if row in [1, 2, 3] and col in [1, 2, 3]:  # Check if row and col are within the board range (1-3)
                if board[row-1][col-1] == " ":  # Adjust index for 0-based Python lists
                    board[row-1][col-1] = player
                    break
                else:
                    print("That spot is already taken. Try again.")
            else:
                print("Invalid input. Please enter row and column numbers between 1 and 3.")
        except ValueError:
            print("Invalid input. Please enter two numbers separated by a space.")

def play_game():
    board = [[" " for _ in range(3)] for _ in range(3)]
    current_player = "X"

    while True:
        print_board(board)
        get_player_input(current_player, board)

        if check_winner(board, current_player):
            print_board(board)
            print(f"Player {current_player} wins!")
            break

        elif check_tie(board):
            print_board(board)
            print("It's a tie!")
            break

        current_player = "O" if current_player == "X" else "X"

    replay = input("Do you want to play again? (yes/no): ").lower()
    if replay == "yes":
        play_game()

play_game()
```

<!-- Project 2: To-Do List -->
<div style="background-color: #f7f7f7; padding: 20px; border-left: 5px solid #4CAF50; margin: 20px 0;">
  <h1 style="color: #388E3C;">Project 2: To-Do List Application ‚úÖ</h1>
  <hr style="border: 1px solid #C8E6C9;">
  
  <h2>Objective</h2>
  <p>Create a simple command-line To-Do list application using Python that allows users to manage their tasks effectively.</p>
  
  <h3>Features:</h3>
  <ul>
    <li>üìù <strong>Add Task:</strong>
      <ul>
        <li>Function: <code>add_task(task)</code></li>
        <li>Stores task as string in list</li>
      </ul>
    </li>
    <li>üëÄ <strong>View Tasks:</strong>
      <ul>
        <li>Function: <code>view_tasks()</code></li>
        <li>Displays tasks with indices</li>
      </ul>
    </li>
    <li>üóëÔ∏è <strong>Delete Task:</strong>
      <ul>
        <li>Function: <code>delete_task(index)</code></li>
        <li>Removes task by index</li>
      </ul>
    </li>
    <li>üîÑ <strong>Program Flow:</strong>
      <ul>
        <li>Interactive menu system</li>
        <li>Input validation</li>
        <li>Exit option</li>
      </ul>
    </li>
  </ul>
</div>

```python
class TodoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, task):
        self.tasks.append(task)
        print(f"Task '{task}' added successfully!")

    def view_tasks(self):
        if not self.tasks:
            print("No tasks in the list!")
            return

        print("\nCurrent Tasks:")
        for i, task in enumerate(self.tasks, 1):
            print(f"{i}. {task}")

    def delete_task(self, index):
        try:
            index = int(index) - 1
            if 0 <= index < len(self.tasks):
                deleted_task = self.tasks.pop(index)
                print(f"Task '{deleted_task}' deleted successfully!")
            else:
                print("Invalid task index!")
        except ValueError:
            print("Please enter a valid number!")

def main():
    todo = TodoList()

    while True:
        print("\n=== To-Do List Menu ===")
        print("1. Add Task")
        print("2. View Tasks")
        print("3. Delete Task")
        print("4. Exit")

        choice = input("\nEnter your choice (1-4): ")

        if choice == '1':
            task = input("Enter the task: ")
            todo.add_task(task)

        elif choice == '2':
            todo.view_tasks()

        elif choice == '3':
            todo.view_tasks()
            if todo.tasks:
                index = input("Enter the task number to delete: ")
                todo.delete_task(index)

        elif choice == '4':
            print("Thank you for using the To-Do List application!")
            break

        else:
            print("Invalid choice! Please try again.")

if __name__ == "__main__":
    main()
```

<!-- Project 3: Robot Path Planning -->
<div style="background-color: #f7f7f7; padding: 20px; border-left: 5px solid #2196F3; margin: 20px 0;">
  <h1 style="color: #1976D2;">Project 3: Robot Path Planning Application ü§ñ</h1>
  <hr style="border: 1px solid #BBDEFB;">
  
  <h2>Problem Statement</h2>
  <p>Develop a Python-based command-line application that simulates a robot navigating through a grid-based environment using A* pathfinding algorithm.</p>
  
  <h3>Requirements:</h3>
  <ul>
    <li>üíª <strong>Libraries Required:</strong>
      <code>numpy</code>, <code>matplotlib</code>, <code>pandas</code>
    </li>
    <li>üî≤ <strong>Environment:</strong> 2D grid representation using NumPy arrays
      <ul>
        <li>Free spaces: 0</li>
        <li>Obstacles: 1</li>
      </ul>
    </li>
    <li>üìç <strong>User Inputs:</strong>
      <ul>
        <li>Grid dimensions (rows and columns)</li>
        <li>Obstacle positions</li>
        <li>Start and end coordinates</li>
      </ul>
    </li>
    <li>üéØ <strong>A* Algorithm Implementation:</strong>
      <ul>
        <li>g: Actual distance from start</li>
        <li>h: Manhattan distance heuristic</li>
        <li>f = g + h: Total cost</li>
      </ul>
    </li>
    <li>üìä <strong>Visualization:</strong> Using matplotlib
      <ul>
        <li>Black cells: Obstacles</li>
        <li>Red dots: Path</li>
        <li>Green: Start</li>
        <li>Blue: End</li>
      </ul>
    </li>
  </ul>
</di

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

class RobotPathPlanner:
    def __init__(self, rows, cols):
        self.rows = rows
        self.cols = cols
        self.grid = np.zeros((rows, cols))

    def is_valid_position(self, x, y):
        return 0 <= x < self.rows and 0 <= y < self.cols

    def add_obstacles(self, obstacles):
        for x, y in obstacles:
            if self.is_valid_position(x, y):
                self.grid[x][y] = 1

    def manhattan_distance(self, p1, p2):
        return abs(p1[0] - p2[0]) + abs(p1[1] - p2[1])

    def get_neighbors(self, current):
        neighbors = []
        for dx, dy in [(0, 1), (1, 0), (0, -1), (-1, 0)]:  # Right, Down, Left, Up
            new_x, new_y = current[0] + dx, current[1] + dy
            if self.is_valid_position(new_x, new_y) and self.grid[new_x][new_y] != 1:
                neighbors.append((new_x, new_y))
        return neighbors

    def find_path(self, start, end):
        if not (self.is_valid_position(start[0], start[1]) and self.is_valid_position(end[0], end[1])):
            return None

        if self.grid[start[0]][start[1]] == 1 or self.grid[end[0]][end[1]] == 1:
            return None

        # Initialize lists for open and closed nodes
        open_list = [(self.manhattan_distance(start, end), 0, start)]  # f_score, g_score, position
        closed_list = set()
        came_from = {start: None}
        g_scores = {start: 0}

        while open_list:
            # Sort open_list by f_score and get the node with lowest f_score
            open_list.sort()
            current_f, current_g, current = open_list.pop(0)

            if current == end:
                # Reconstruct path
                path = []
                while current is not None:
                    path.append(current)
                    current = came_from[current]
                return path[::-1]

            closed_list.add(current)

            for neighbor in self.get_neighbors(current):
                if neighbor in closed_list:
                    continue

                tentative_g = g_scores[current] + 1

                # Check if neighbor is in open list and if this path is better
                if neighbor not in g_scores or tentative_g < g_scores[neighbor]:
                    came_from[neighbor] = current
                    g_scores[neighbor] = tentative_g
                    f_score = tentative_g + self.manhattan_distance(neighbor, end)

                    # Update or add to open list
                    open_list = [(f, g, pos) for f, g, pos in open_list if pos != neighbor]
                    open_list.append((f_score, tentative_g, neighbor))

        return None

    def visualize_path(self, path, start, end):
        plt.figure(figsize=(10, 10))
        plt.imshow(self.grid, cmap='binary')

        if path:
            path = np.array(path)
            plt.plot(path[:, 1], path[:, 0], 'r.-', label='Path')

        plt.plot(start[1], start[0], 'go', label='Start', markersize=15)
        plt.plot(end[1], end[0], 'bo', label='End', markersize=15)
        plt.grid(True)
        plt.legend()
        plt.show()

    def display_grid_data(self):
        return pd.DataFrame(self.grid)

def get_valid_input(prompt, max_row, max_col):
    while True:
        try:
            x, y = map(int, input(prompt).split())
            if 0 <= x < max_row and 0 <= y < max_col:
                return x, y
            else:
                print(f"Coordinates must be within bounds: 0-{max_row-1} for row and 0-{max_col-1} for column")
        except ValueError:
            print("Please enter two numbers separated by a space")

def main():
    # Get user input for grid size
    while True:
        try:
            rows = int(input("Enter number of rows: "))
            cols = int(input("Enter number of columns: "))
            if rows > 0 and cols > 0:
                break
            print("Rows and columns must be positive numbers")
        except ValueError:
            print("Please enter valid numbers")

    planner = RobotPathPlanner(rows, cols)

    # Get number of obstacles
    while True:
        try:
            n_obstacles = int(input("Enter number of obstacles: "))
            if n_obstacles >= 0:
                break
            print("Number of obstacles must be non-negative")
        except ValueError:
            print("Please enter a valid number")

    # Get obstacle coordinates
    print(f"Enter obstacle coordinates (row col) one per line (0-{rows-1} for row, 0-{cols-1} for col):")
    obstacles = []
    for _ in range(n_obstacles):
        x, y = get_valid_input("", rows, cols)
        obstacles.append((x, y))
    planner.add_obstacles(obstacles)

    # Get start and end positions
    print(f"\nValid coordinates are: 0-{rows-1} for row and 0-{cols-1} for column")
    start_x, start_y = get_valid_input("Enter start position (row col): ", rows, cols)
    end_x, end_y = get_valid_input("Enter end position (row col): ", rows, cols)

    # Display initial grid
    print("\nInitial Grid:")
    print(planner.display_grid_data())

    # Find and visualize path
    path = planner.find_path((start_x, start_y), (end_x, end_y))
    if path:
        print("\nPath found!")
        print("Path coordinates:", path)
        planner.visualize_path(path, (start_x, start_y), (end_x, end_y))
    else:
        print("\nNo valid path exists!")

if __name__ == "__main__":
    main()
```

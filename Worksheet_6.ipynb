```python
#q1
import numpy as np
import matplotlib.pyplot as plt

marks_data = np.array([
    [85, 78, 92, 88],  # Arin
    [79, 82, 74, 90],  # Aditya
    [90, 85, 89, 92],  # Chirag
    [66, 75, 80, 78],  # Gurleen
    [70, 68, 75, 85]   # Kunal
])

students = ['Arin', 'Aditya', 'Chirag', 'Gurleen', 'Kunal']
subjects = ['Math', 'Physics', 'Chemistry', 'English']

total_marks = np.sum(marks_data, axis=1)
average_marks = np.mean(marks_data, axis=1)

subject_wise_average = np.mean(marks_data, axis=0)

top_performer_index = np.argmax(total_marks)
bottom_performer_index = np.argmin(total_marks)

top_performer = students[top_performer_index]
bottom_performer = students[bottom_performer_index]

print(f"Total marks for each student: {dict(zip(students, total_marks))}")
print(f"Average marks for each student: {dict(zip(students, average_marks))}")
print(f"Subject-wise average marks: {dict(zip(subjects, subject_wise_average))}")
print(f"Top performer: {top_performer} with {total_marks[top_performer_index]} marks")
print(f"Bottom performer: {bottom_performer} with {total_marks[bottom_performer_index]} marks")

plt.figure(figsize=(10, 5))
plt.bar(students, total_marks, color='skyblue')
plt.xlabel('Students')
plt.ylabel('Total Marks')
plt.title('Total Marks of Students')
plt.show()

plt.figure(figsize=(10, 5))
plt.bar(subjects, subject_wise_average, color='lightcoral')
plt.xlabel('Subjects')
plt.ylabel('Average Marks')
plt.title('Average Marks for Each Subject')
plt.show()
```

```python
#q2
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

time = np.array([0, 1, 2, 3, 4, 5])
velocity = np.array([2, 3.1, 7.9, 18.2, 34.3, 56.2])

def quadratic(t, a, b, c):
    return a*t**2 + b*t + c

params, _ = curve_fit(quadratic, time, velocity)

fitted_velocity = quadratic(time, *params)

plt.plot(time, velocity, 'o', label='Data')
plt.plot(time, fitted_velocity, '-', label='Fitted Curve')
plt.xlabel('Time (seconds)')
plt.ylabel('Velocity (m/s)')
plt.legend()
plt.show()
```

```python
#q3
import numpy as np
from scipy import stats
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

years = np.array([2000, 2005, 2010, 2015, 2020])
population = np.array([50, 55, 70, 80, 90])

correlation, _ = stats.pearsonr(years, population)
print(f"Pearson's correlation coefficient: {correlation}")

interp_func = interp1d(years, population, kind='linear')

population_2008 = interp_func(2008)
print(f"Estimated population in 2008: {population_2008}")

plt.plot(years, population, 'o', label='Data')
plt.plot(2008, population_2008, 'ro', label='Estimate (2008)')
plt.plot(np.linspace(2000, 2020, 100), interp_func(np.linspace(2000, 2020, 100)), '-', label='Interpolation')
plt.xlabel('Year')
plt.ylabel('Population (in thousands)')
plt.legend()
plt.show()
```

```python
#q4
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

def polynomial(x):
    return 3*x**3 - 5*x**2 + 2*x - 8

roots = np.roots([3, -5, 2, -8])
print(f"Roots of the polynomial: {roots}")

x = np.linspace(-3, 3, 100)
y = polynomial(x)

plt.plot(x, y, label='Polynomial')
plt.axhline(0, color='black', linewidth=0.5)
plt.scatter(roots, polynomial(roots), color='red', label='Roots')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.legend()
plt.title('Polynomial and its Roots')
plt.show()
```

```python
#q5
import time
import random
import string
import matplotlib.pyplot as plt

# Function to generate random file of given size in MB
def generate_file(filename, size_mb):
    size = size_mb * 1024 * 1024  # Size in bytes
    with open(filename, 'w') as f:
        f.write(''.join(random.choices(string.ascii_lowercase, k=size)))

# Function to convert file content to uppercase and measure time taken
def convert_to_upper(filename):
    start_time = time.time()
    with open(filename, 'r') as f:
        data = f.read().upper()  # Read the file and convert to uppercase
    end_time = time.time()
    return end_time - start_time  # Return time taken

# File sizes in MB
file_sizes = [200, 400, 600, 800, 1000]

# List to store time taken for each file size
time_taken = []

# Generate files and measure time taken for conversion for each file size
for size in file_sizes:
    filename = f"file_{size}mb.txt"
    print(f"Generating {size}MB file...")
    generate_file(filename, size)

    print(f"Converting {size}MB file to upper case...")
    elapsed_time = convert_to_upper(filename)
    time_taken.append(elapsed_time)
    print(f"Time taken for {size}MB file: {elapsed_time:.2f} seconds")

# Plot the results
plt.figure(figsize=(10, 5))
plt.plot(file_sizes, time_taken, marker='o', linestyle='-', color='b')
plt.title('Time Taken to Convert Files to Upper Case')
plt.xlabel('File Size (MB)')
plt.ylabel('Time Taken (seconds)')
plt.grid(True)
plt.show()
```

```python
#q6
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt

def f(x):
    return x**4 - 3*x**3 + 2

result = minimize(f, x0=0)
print(f"Local minima at: {result.x}")

x = np.linspace(-2, 3, 100)
y = f(x)

plt.plot(x, y, label='f(x)')
plt.scatter(result.x, f(result.x), color='red', label='Local Minima')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Function and Local Minima')
plt.show()
```
